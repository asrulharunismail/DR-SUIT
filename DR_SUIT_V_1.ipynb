```python
#@title BEES ALGORITHM's code

Scout_Bees= 25
Maximum_Iteration= 25

import numpy as np
from numpy import random
import math
import pandas as pd
from scipy.spatial.distance import cdist, pdist, squareform
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2, radians

n = int(Scout_Bees)
m = int(0.5*n)
e = int(0.1*n)
nep = int(n)
nsp = int(0.5*n)
ngh = float(0.1)
MaxIt = int(Maximum_Iteration)
s = float(0.9)
sc = int(Drone_Swarm)
long = float(Longitude_of_onshore_base)
lat = float(Latitude_of_onshore_base)
tier=int(Tier)
#VSpeed = float(Vessel_Speed)
#vesselSpeed = VSpeed*60/1000
droneSpeed=float(Drone_Speed)
inspectionTime=int(Inspection_Time)
#chargingTime=int(Charging_Time)  ####
#timeWindow=int(Time_Windows)
extraVessel=0

maxIt = MaxIt  # Maximum Number of Iterations
nScoutBee = n  # n = Number of Scout Bees
nSelectedSite = m  # m = Number of Selected Sites
nEliteSite = e  # e = Number of Selected Elite Sites
nSelectedSiteBee = nsp  # nsp = Number of Selected Recruited Bees for Selected (m-e) Sites
nEliteSiteBee = nep  # nep = Number of Recruited Bees for Elite Sites
shrink = s

def updateMat (d):
    clustNo = d.shape[1]
    nodeNo = d.shape[0]
    monitoring = np.zeros((1,clustNo))
    monitorNo = np.floor(nodeNo/clustNo)
    remNode= nodeNo%clustNo
    clustPerNode=np.zeros((1,nodeNo))
    A=np.array(d)
    B=np.array(d)
    ANo= nodeNo-remNode
    BNo= remNode
    clustDis = np.zeros((1,clustNo))
    MINI=[]
    MINI2=[]
    for i in range(ANo):
        MINI=np.amin(A)
        id1 = np.argwhere(A==MINI)
        node = id1[0][0] # cluster terpilih
        A[node,:] = np.inf
        B[node,:] = np.inf
        clust = id1[0][1]
        monitoring[0][clust]= monitoring[0][clust]+1
        clustDis[0][clust]=clustDis[0][clust]+MINI
        F=np.argwhere(monitoring[0]==monitorNo)
        A[:,F] = np.inf
        clustPerNode[0][node]=clust

    if remNode > 0:

        for i in range(BNo):
            MINI2=np.amin(B)
            id2 = np.argwhere(B==MINI2)
            node2 = id2[0][0] # cluster terpilih
            B[node2,:] = np.inf  # titik terpilih
            clust = id2[0][1]
            B[:,clust] = np.inf # update mat clust terpilih
            monitoring[0][clust]= monitoring[0][clust]+1
            clustDis[0][clust]=clustDis[0][clust]+MINI2
            clustPerNode[0][node2] =clust

    clustPerNode=np.array(clustPerNode).flatten()
    clustPerNode= clustPerNode.transpose()
    dis= np.array(clustDis).flatten()
    return clustPerNode, dis

def ND (k,X,Z):
    C=[]
    nodeC=[]
    for i in range(k):
        C.append([])
        nodeC.append([])
    CP=np.zeros((k,2),float)

    for m in range (k):
        C[m] = np.argwhere(Z==m)
        nodeC[m] = X[C[m],:]
        nodeC[m] = np.array(nodeC[m]).reshape(-1,2)
        CP[m,0] = nodeC[m][:,0].mean(axis=0)
        CP[m,1] = nodeC[m][:,1].mean(axis=0)
    CP=np.array(CP)
    return CP, nodeC

def Foraging(x,ngh,Vmx,Vmn):
    x=np.array(x)
    nVar = x.shape[0]
    k = np.random.randint(0, nVar)
    L = round(np.random.random())
    y = x
    y=np.array(y)
    if L==0:
        y[k,0] = x[k,0] + (np.random.uniform(-ngh, ngh)*(Vmx[0]-Vmn[0]))
        y[:,0][y[:,0]>Vmx[0]]=Vmx[0]
        y[:,0][y[:,0]<Vmn[0]]=Vmn[0]
    else:
        y[k,1] = x[k,1] + (np.random.uniform(-ngh, ngh)*(Vmx[1]-Vmn[1]))
        y[:,1][y[:,1]>Vmx[1]]=Vmx[1]
        y[:,1][y[:,1]<Vmn[1]]=Vmn[1]
    return y

class Bee:
    def __init__(self,Position,Cost,Out,Node,Seq,Nod):
        self.Position = Position
        self.Cost = Cost
        self.Out = Out
        self.Node = Node
        self.Seq = Seq
        self.Nod = Nod

class prob:
    def __init__(self,tipe,noCluster,depo):
        self.tipe = tipe
        self.noCluster = noCluster
        self.depo = depo

class VS:
    def __init__(self,ori,des,vessel,Nth,batch,Tier,leng,dur,esT,efT,insp,ch,esI,efI):
        self.ori = ori
        self.des = des
        self.vessel = vessel
        self.Nth = Nth
        self.batch = batch
        self.Tier = Tier
        self.leng = leng
        self.dur = dur
        self.esT = esT
        self.efT = efT
        self.insp = insp
        self.ch = ch
        self.esI = esI
        self.efI = efI

class DS:
    def __init__(self,ori1,ori2,des1,des2,dis,insp,esI,efI):
        self.ori1 = ori1
        self.ori2 = ori2
        self.des1 = des1
        self.des2 = des2
        self.dis = dis
        self.insp = insp
        self.esI = esI
        self.efI = efI

def GeoDis(a,b,c,d):
    Long1 = radians(a)
    Lat1 = radians(b)
    Long2 = radians(c)
    Lat2 = radians(d)
    # approximate radius of earth in km
    R = 6373.0

    dLat = Lat2 - Lat1
    dLong = Long2 - Long1

    a = sin(dLong / 2)**2 + cos(Long1) * cos(Long2) * sin(dLat / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

def TwoOpt (D):
    n = D.shape[0]
    tour = np.random.permutation(n)
    Zmin = -1
    while Zmin < 0 :
        Zmin = 0
        i=-1
        b=tour[-1]
        while i < n-3:
            a = b
            i += 1
            b = tour[i]
            Dab = D[a,b]
            j = i +1
            d = tour[j]
            while j < n-1:
                c = d
                j += 1
                d = tour[j]
                Z = (D[a,c]-D[c,d]) + D[b,d] - Dab
                if Z < Zmin:
                    Zmin = Z
                    imin = i
                    jmin = j
        if Zmin < 0:
            tour=list(tour)
            tour[imin:jmin:1]= reversed(tour[imin:jmin:1])
    L=0
    for i in range (n):
        if i< n-1:
          L += D[tour[i],tour[i+1]]
        else:
          L += D[tour[i],tour[0]]
    return tour, L

def Cost (a,x,data):
    if a.tipe==1:
        d=cdist(data,x)
        idmin= updateMat(d)
        dmin=np.array(idmin[1])
        z=dmin.sum()
        out = np.array(idmin[0])
        seq=0
        nod=0
    else:
        depoPoint=a.depo
        d=cdist(data,x)
        idmin= updateMat(d)
        clustNo=x.shape[0]
        tot=[]
        seq=[]
        NOD=[]
        for i in range(clustNo):
            tot.append([])
            seq.append([])
            NOD.append([])
        for i in range(clustNo):
            node=[]
            idd=np.argwhere(idmin[0]==i)
            node=data[idd]
            node=np.array(node).reshape(-1,2)
            node=np.vstack((node, depoPoint))
            matDuration=squareform(pdist(node));
            if sum(sum(matDuration))==0:
                out=0
                seq=0
                z=0
                nod=0
            else:
                TL = TwoOpt(matDuration)

            tot[i]= TL[1]
            seq[i]= TL[0]
            NOD[i]= node
        z=sum(tot)
        out = np.array(idmin[0])
        seq = np.array(seq)
        nod = np.array(NOD)
    return z,out,seq,nod

def BEES(data,problem,maxIt,nScoutBee,nSelectedSite,nEliteSite,nSelectedSiteBee,nEliteSiteBee,ngh,shrink):
    varMin = np.amin(data,0)     ####
    varMax = np.amax(data,0)     ####
    k=problem.noCluster

    # Initialization
    Patch=[]
    for i in range(nScoutBee):
        Patch.append([])
        position = data[np.random.randint(0,data.shape[0],size=(1,k)),:]
        position = np.array(position).reshape(-1,2)
        Patch[i]=Bee(position,Cost(problem,position,data)[0],[],[],[],[])

    Patch.sort(key=lambda Patch: Patch.Cost, reverse=False)

    BestSol = ([],math.inf,0)
    BestSol = Patch[0]

    BestCost = np.zeros([maxIt,1])
    BestPos = []

    for i in range(maxIt):
        BestPos.append([])

    newbee=Bee([],[],[],[],[],[])


    # Main Loop
    for it in range(maxIt):

        # Elite Sites
        for i in range(nEliteSite):

            bestnewbee=Bee([],math.inf,[],[],[],[])

            for j in range(nEliteSiteBee):
                newbee.Position = Foraging(Patch[i].Position,ngh,varMax,varMin)
                newbee.Position = np.array(newbee.Position).reshape(-1,2)
                COUT=Cost(problem,newbee.Position,data)
                newbee.Cost = COUT[0]
                newbee.Out = COUT[1]
                NOD = ND(problem.noCluster,data,newbee.Out)
                newbee.Node = NOD[1]
                newbee.Position = NOD[0]
                newbee.Seq = COUT[2]
                newbee.Nod = COUT[3]
                if newbee.Cost < bestnewbee.Cost:
                    bestnewbee.Cost = newbee.Cost
                    bestnewbee.Position = newbee.Position
                    bestnewbee.Out = newbee.Out
                    bestnewbee.Node = newbee.Node
                    bestnewbee.Seq = newbee.Seq
                    bestnewbee.Nod = newbee.Nod

            if bestnewbee.Cost < Patch[i].Cost:
                Patch[i].Cost = bestnewbee.Cost
                Patch[i].Position = bestnewbee.Position
                Patch[i].Out = bestnewbee.Out
                Patch[i].Node = bestnewbee.Node
                Patch[i].Seq = bestnewbee.Seq
                Patch[i].Nod = bestnewbee.Nod

        # Selected Non-Elite Sites
        for i in range(nEliteSite,nSelectedSite):

            bestnewbee=Bee([],math.inf,[],[],[],[])

            for j in range(nSelectedSiteBee):
                newbee.Position = Foraging(Patch[i].Position,ngh,varMax,varMin)
                newbee.Position = np.array(newbee.Position).reshape(-1,2)
                COUT=Cost(problem,newbee.Position,data)
                newbee.Cost = COUT[0]
                newbee.Out = COUT[1]
                NOD = ND(problem.noCluster,data,newbee.Out)
                newbee.Node = NOD[1]
                newbee.Position = NOD[0]
                newbee.Seq = COUT[2]
                newbee.Nod = COUT[3]
                if newbee.Cost < bestnewbee.Cost:
                    bestnewbee.Cost = newbee.Cost
                    bestnewbee.Position = newbee.Position
                    bestnewbee.Out = newbee.Out
                    bestnewbee.Node = newbee.Node
                    bestnewbee.Seq = newbee.Seq
                    bestnewbee.Nod = newbee.Nod

            if bestnewbee.Cost < Patch[i].Cost:
                Patch[i].Cost = bestnewbee.Cost
                Patch[i].Position = bestnewbee.Position
                Patch[i].Out = bestnewbee.Out
                Patch[i].Node = bestnewbee.Node
                Patch[i].Seq = bestnewbee.Seq
                Patch[i].Nod = bestnewbee.Nod

        # Non - Selected Sites
        for i in range(nSelectedSite,nScoutBee):
            XX = np.random.uniform(varMin[0],varMax[0],size=(k,1))
            YY = np.random.uniform(varMin[1],varMax[1],size=(k,1))
            position = np.hstack((XX,YY))
            position = np.array(position).reshape(-1,2)
            COUT=Cost(problem,position,data)
            NOD = ND(problem.noCluster,data,COUT[1])
            Patch[i]=Bee(position,COUT[0],COUT[1],NOD[1],COUT[2],COUT[3])
            Patch[i].Position = NOD[0]

        Patch=sorted(Patch, key=lambda Patch: Patch.Cost, reverse=False)

        BestSol.Cost = Patch[0].Cost
        BestSol.Position = Patch[0].Position
        BestSol.Out = Patch[0].Out
        BestSol.Node = Patch[0].Node
        BestSol.Seq = Patch[0].Seq
        BestSol.Nod = Patch[0].Nod

        BestCost[it] = BestSol.Cost
        BestPos [it] = BestSol.Position

        print(['Iteration ' + str(it) + ': calculating......' ]) #+ str(BestCost[it]) ])

        ngh = shrink * ngh
    return BestSol

def PLT (k,p,n):
    x = np.array(p[:,0])
    y = np.array(p[:,1])
    plt.scatter(y, x, color = 'black', marker='o')
    for i in range(k):
        x = np.array(n[i][:,0])
        y = np.array(n[i][:,1])
        plt.scatter(y, x, color = [random.rand(),random.rand(),random.rand()], marker='x')
    return plt.show()

def PLT2 (k1,k2,p,n,rnode):
    x = np.array(p[:,0])
    y = np.array(p[:,1])
    plt.scatter(y, x, color = 'black', marker='o')
    for i in range(k1):
        x = np.array(n[i][:,0])
        y = np.array(n[i][:,1])
        plt.scatter(y, x, color = [random.rand(),random.rand(),random.rand()], marker='x')

    for i in range(k2):
        x = np.array(rnode[i][:,0])
        y = np.array(rnode[i][:,1])
        plt.plot(y, x)
    plt.savefig('ROUTING.pdf')
    return plt.show()

```

```python
#@title CLUSTERING ("Press Play to Replay")
import copy

DATA=np.array(DATA)
data1=DATA
problem1=prob([],[],[])
problem1.tipe=1
problem1.noCluster=int(np.ceil(DATA.shape[0]/sc))
problem1.depo=[long, lat]
problem1.xmin=min(min(DATA[:,0]),long)
problem1.xmax=max(max(DATA[:,1]),long)

CLUSTERING= BEES(data1,problem1,maxIt,nScoutBee,nSelectedSite,nEliteSite,nSelectedSiteBee,nEliteSiteBee,ngh,shrink)

from pylab import rcParams
rcParams['figure.figsize'] =20,15
P1=PLT(problem1.noCluster,CLUSTERING.Position,CLUSTERING.Node)

stopVesselPoint = copy.copy (CLUSTERING.Position)

inspectionNodes = CLUSTERING.Node

## MOVE TO MIDDLE POINT
for i in range(CLUSTERING.Position.shape[0]):
    dis=cdist(np.array(stopVesselPoint[i,:]).reshape(-1,2),DATA)
    disMin = sorted(dis, key=None, reverse=False)
    idx=np.argsort(dis)
    idx=np.array(idx).flatten()
    stopVesselPoint[i,:] = (DATA[idx[0],:]+DATA[idx[1],:])/2

safePoints = (CLUSTERING.Position + stopVesselPoint)/2

#P2=PLT(problem1.noCluster,CLUSTERING.Position,CLUSTERING.Node)
#P2=PLT(problem1.noCluster,safePoints, CLUSTERING.Node)

CASE=np.zeros((CLUSTERING.Position.shape[0],inspectionNodes[0].shape[0]))
DIS=np.zeros((CLUSTERING.Position.shape[0],inspectionNodes[0].shape[0]))
for i in range(CLUSTERING.Position.shape[0]):
    for j in range (inspectionNodes[i].shape[0]):
        CASE[i,j]= (alt_km/Vacs)+(alt_km/Vdecs)+(GeoDis(stopVesselPoint[i,0],stopVesselPoint[i,1],inspectionNodes[i][j,0],inspectionNodes[i][j,1])/droneSpeed*2)+inspectionTime
        DIS[i,j]= GeoDis(stopVesselPoint[i,0],stopVesselPoint[i,1],inspectionNodes[i][j,0],inspectionNodes[i][j,1])

droneSchedule=[]
for i in range (problem1.noCluster):
    droneSchedule.append([])

for i in range (problem1.noCluster):
    cc=CASE[i,:]
    cc=np.delete(cc,np.where(cc==0)) ###
    droneSchedule[i].append(np.transpose(cc))  ## CASE[i,:]=cc
    droneSchedule[i].append(CLUSTERING.Position[i,0])
    droneSchedule[i].append(CLUSTERING.Position[i,1])
    droneSchedule[i].append(CLUSTERING.Node[i])
    droneSchedule[i].append(max(CASE[i,:]))
    droneSchedule[i].append(DIS[i,:])

droneSchedule=np.array(droneSchedule)
```

```python
#@title ROUTING AND SCHEDULING (If you want to reroute the vessel mission plan without restarting the clustering problem, you must update the number of vessel (above this section -INPUT 2-) and click the play button in this section.)
vesselPoint = CLUSTERING.Position
data2 = vesselPoint
problem2=prob([],[],[])
problem2.tipe=2
problem2.noCluster=int(number_of_Vessel)
problem2.depo=[long, lat]
problem2.xmin=min(min(DATA[:,0]),long)
problem2.xmax=max(max(DATA[:,1]),long)

ROUTING= BEES(data2,problem2,maxIt,nScoutBee,nSelectedSite,nEliteSite,nSelectedSiteBee,nEliteSiteBee,ngh,shrink)

routeSeq=ROUTING.Seq
routeNode=[]
for i in range(problem2.noCluster):
    routeNode.append([])
for i in range(problem2.noCluster):
    while routeSeq[i][0]!=max(routeSeq[i]):
        routeSeq[i]=np.roll(routeSeq[i],-1)
    routeNode[i]=ROUTING.Nod[i][routeSeq[i],:]

#P3=PLT2(problem1.noCluster,problem2.noCluster,CLUSTERING.Position,CLUSTERING.Node,routeNode)

vesselSchedule=[]
for i in range (routeSeq.shape[0]):
    vesselSchedule.append(VS([],[],[],[],[],[],[],[],[],[],[],[],[],[]))

droneSchedule=np.array(droneSchedule)

for i in range(routeSeq.shape[0]):
    vesselSchedule[i].ori=routeNode[i]
    vesselSchedule[i].des=np.roll(routeNode[i],-1,axis=0)
    for j in range (routeNode[i].shape[0]):
        vesselSchedule[i].vessel.append(i)
        if j==routeNode[i].shape[0]-1:
            vesselSchedule[i].Nth.append(0)
        else:
            vesselSchedule[i].Nth.append(j+1)
        vesselSchedule[i].batch.append(np.ceil((j+1)/tier))
        if (j+1)%tier==0:
            vesselSchedule[i].Tier.append(tier)
        else:
            vesselSchedule[i].Tier.append((j+1)%tier)
        vesselSchedule[i].leng.append(GeoDis(vesselSchedule[i].ori[j,0],vesselSchedule[i].ori[j,1],vesselSchedule[i].des[j,0],vesselSchedule[i].des[j,1]))
        vesselSchedule[i].dur.append(GeoDis(vesselSchedule[i].ori[j,0],vesselSchedule[i].ori[j,1],vesselSchedule[i].des[j,0],vesselSchedule[i].des[j,1])/vesselSpeed)
        if j==0:
            vesselSchedule[i].esT.append(0)
            vesselSchedule[i].efT.append(vesselSchedule[i].esT[j]+vesselSchedule[i].dur[j])
            AA1=np.argwhere(droneSchedule[:,1]==routeNode[i][j+1,0])
            AA2=np.argwhere(droneSchedule[:,2]==routeNode[i][j+1,1])
            AA=np.intersect1d(AA1,AA2)
            vesselSchedule[i].insp.append(droneSchedule[AA,4])
            vesselSchedule[i].ch.append((vesselSchedule[i].insp[-1]/layer)/flyingDuration*init_chargingTime) ####
            vesselSchedule[i].esI.append(np.zeros(tier))
            vesselSchedule[i].esI[j][0]=vesselSchedule[i].efT[j]
            vesselSchedule[i].efI.append(np.zeros(tier))
            vesselSchedule[i].efI[j][0]=vesselSchedule[i].esI[j][0]+vesselSchedule[i].insp[j]
        elif j==routeNode[i].shape[0]-1:
            vesselSchedule[i].esT.append(max(max(vesselSchedule[i].esI[j-1][:]),max(vesselSchedule[i].efI[j-1][:])))
            vesselSchedule[i].efT.append(vesselSchedule[i].esT[j]+vesselSchedule[i].dur[j])
            vesselSchedule[i].insp.append(0)
            vesselSchedule[i].ch.append(0)
            vesselSchedule[i].esI.append(np.zeros(tier))
            vesselSchedule[i].efI.append(np.zeros(tier))
        else:
            vesselSchedule[i].esT.append(max(max(vesselSchedule[i].esI[j-1][:]),max(vesselSchedule[i].efI[j-1][:])))
            vesselSchedule[i].efT.append(vesselSchedule[i].esT[j]+vesselSchedule[i].dur[j])
            AA1=np.argwhere(droneSchedule[:,1]==routeNode[i][j+1,0])
            AA2=np.argwhere(droneSchedule[:,2]==routeNode[i][j+1,1])
            AA=np.intersect1d(AA1,AA2)
            vesselSchedule[i].insp.append(droneSchedule[AA,4])
            vesselSchedule[i].ch.append((vesselSchedule[i].insp[-1]/layer)/flyingDuration*init_chargingTime)  ####
            vesselSchedule[i].esI.append(np.zeros(tier))
            vesselSchedule[i].efI.append(np.zeros(tier))
            if vesselSchedule[i].batch[j]==1:
                vesselSchedule[i].esI[j][vesselSchedule[i].Tier[j]-1]= vesselSchedule[i].efT[j]
                vesselSchedule[i].efI[j][vesselSchedule[i].Tier[j]-1]= vesselSchedule[i].esI[j][vesselSchedule[i].Tier[j]-1]+vesselSchedule[i].insp[j]
            else:
                vesselSchedule[i].esI[j][vesselSchedule[i].Tier[j]-1]=max(vesselSchedule[i].efT[j],max(vesselSchedule[i].esI[j-tier] + (vesselSchedule[i].insp[j-tier])/layer)+vesselSchedule[i].ch[j])
                vesselSchedule[i].efI[j][vesselSchedule[i].Tier[j]-1]=vesselSchedule[i].esI[j][vesselSchedule[i].Tier[j]-1]+vesselSchedule[i].insp[j]

for i in range (len(vesselSchedule)):
    dt=[]
    dt = {'Original_Long': vesselSchedule[i].ori[:,0],
          'Original_Lat': vesselSchedule[i].ori[:,1],
          'Destination_Long': vesselSchedule[i].des[:,0],
          'Destination_Lat': vesselSchedule[i].des[:,1],
          'nth Vessel': np.array(vesselSchedule[i].vessel)+1,
          'nth halt': np.array(vesselSchedule[i].Nth),
          'Batch': np.array(vesselSchedule[i].batch),
          'Tier': np.array(vesselSchedule[i].Tier),
          'Travel distance (km)': np.array(vesselSchedule[i].leng),
          'Travel duration (min)': np.array(vesselSchedule[i].dur),
          'Vessel Start': np.array(vesselSchedule[i].esT),
          'Vessel Stop': np.array(vesselSchedule[i].efT),
          'Inspection duration + dr-travel time (min)': vesselSchedule[i].insp,
          'charging time (min)': np.array(vesselSchedule[i].ch),
          }

    # Convert the dictionary into DataFrame
    exec(f'vTable_{i+1} = pd.DataFrame(dt)')
    zz1=np.array(vesselSchedule[i].esI)
    zz2=np.array(vesselSchedule[i].efI)
    for j in range(tier):
        esI=zz1[:,j]
        efI=zz2[:,j]
        eval("vTable_"+str(i+1)) ['Start Inspection tier-'+str(j+1)] = esI
        eval("vTable_"+str(i+1)) ['Finished Inspection tier-'+str(j+1)] = efI

vv=[]
for i in range(len(vesselSchedule)):
    vv.append(eval("vTable_"+str(i+1)))
vTable=pd.concat(vv)

dTable=[]
for i in range (droneSchedule.shape[0]):
    dTable.append(DS([],[],[],[],[],[],[],[]))
    n=droneSchedule[i][0].shape[0]
    ValD=list(droneSchedule[i][0])
    ValD.sort(reverse=True)
    ValDis=list(droneSchedule[i][5])
    ValDis.sort(reverse=True)
    idD=(-droneSchedule[i][0]).argsort()
    dTable[i].ori1=np.array([droneSchedule[i][1]]*n).flatten()
    dTable[i].ori2=np.array([droneSchedule[i][2]]*n).flatten()
    dTable[i].des1=droneSchedule[i][3][idD][:,0]
    dTable[i].des2=droneSchedule[i][3][idD][:,1]
    dTable[i].dis=np.array(ValDis).flatten()
    dTable[i].insp=np.array(ValD).flatten()
    r_d = vTable[(vTable['Destination_Long'] == droneSchedule[i][1]) & (vTable['Destination_Lat'] == droneSchedule[i][2])]
    dTable[i].esI = np.array([r_d.iloc[:,14+((r_d.Tier-1)*2)]]*n).flatten()
    dTable[i].efI = dTable[i].esI + dTable[i].insp

for i in range(len(routeNode)):
    for j in range(1,len(routeNode[i])):
        idDr = np.where((droneSchedule[:,1] == routeNode[i][j][0]) & (droneSchedule[:,2] == routeNode[i][j][1]))[0]
        idDr = int(idDr)
        dT=[]
        dT = {'Original_Long': dTable[idDr].ori1,
              'Original_Lat': dTable[idDr].ori2,
              'Destination_Long': dTable[idDr].des1,
              'Destination_Lat': dTable[idDr].des2,
              'Distance (km)':dTable[idDr].dis[dTable[idDr].dis!=0],
              'Inspection duration + dr-travel time (min)': dTable[idDr].insp,
              'Drone Start': dTable[idDr].esI,
              'Drone Return': dTable[idDr].efI,
              'Vessel_HaltPoint': str(i+1)+"-"+str(j),
              }
        exec(f'DrTable_{i+1}_{j} = pd.DataFrame(dT)')
        if layer ==2:
            eval("DrTable_"+str(i+1)+"_"+str(j)) ['Start Dr layer-1'] = dTable[idDr].esI
            eval("DrTable_"+str(i+1)+"_"+str(j)) ['Finished Dr layer-1'] = dTable[idDr].esI+(dTable[idDr].insp/2)
            eval("DrTable_"+str(i+1)+"_"+str(j)) ['Start Dr layer-2'] = dTable[idDr].esI+(dTable[idDr].insp/2)
            eval("DrTable_"+str(i+1)+"_"+str(j)) ['Finished Dr layer-2'] = dTable[idDr].efI

dd=[]
for i in range(len(routeNode)):
    for j in range(1,len(routeNode[i])):
        dd.append(eval("DrTable_"+str(i+1)+"_"+str(j)))
DrTable=pd.concat(dd)

col=vTable["Vessel Stop"]
print("Congratulation the Calculation has been completed!!!")
print("Your inspection will be finished after: " + str(col.max()) +" minutes")
print("Using "+str(number_of_Vessel)+" vessels, with speed " +str(Vessel_Speed)+ " m/s, each vessel carried " + str(int(Drone_Swarm)*int(Tier)*layer)+" drones")
print("with drone speed: "+str(Drone_Speed)+" m/s, inspection time: " +str(Inspection_Time)+" minutes")
print("For the detail please access the file1 and file2.csv")
```
